```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_rel
```

```python
data = pd.read_csv('LDP.csv')
```

```python
data.head()
```

```python
manfaat_stats = data[['Manfaat Ekspektasi', 'Manfaat Realitas']].describe()

kontribusi_stats = data[['Kontribusi Ekspektasi', 'Kontribusi Realita']].describe()

kinerja_stats = data[['Kinerja Ekspektasi', 'Kinerja Realita']].describe()
```

```python
print("Statistik Deskriptif - Kategori Manfaat:")
print(manfaat_stats)

print("\nStatistik Deskriptif - Kategori Kontribusi:")
print(kontribusi_stats)

print("\nStatistik Deskriptif - Kategori Kinerja:")
print(kinerja_stats)
```

```python
# Histogram Kategori Manfaat
plt.figure(figsize=(10, 6))
sns.histplot(data['Manfaat Ekspektasi'], label='Ekspektasi', kde=True)
sns.histplot(data['Manfaat Realitas'], label='Realita', kde=True)
plt.title('Perbandingan Ekspektasi dan Realita - Manfaat')
plt.legend()
plt.show()
```

```python
# Histogram Kategori Kontribusi
plt.figure(figsize=(10, 6))
sns.histplot(data['Kontribusi Ekspektasi'], label='Ekspektasi', kde=True)
sns.histplot(data['Kontribusi Realita'], label='Realita', kde=True)
plt.title('Perbandingan Ekspektasi dan Realita - Manfaat')
plt.legend()
plt.show()
```

```python
# Histogram Kategori Kinerja
plt.figure(figsize=(10, 6))
sns.histplot(data['Kinerja Ekspektasi'], label='Ekspektasi', kde=True)
sns.histplot(data['Kinerja Realita'], label='Realita', kde=True)
plt.title('Perbandingan Ekspektasi dan Realita - Manfaat')
plt.legend()
plt.show()
```

```python
# Uji t-test Manfaat
t_stat_Manfaat, p_value_Manfaat = ttest_rel(data['Manfaat Ekspektasi'], data['Manfaat Realitas'])

# Uji t-test Kontribusi
t_stat_Kontribusi, p_value_Kontribusi = ttest_rel(data['Kontribusi Ekspektasi'], data['Kontribusi Realita'])

# Uji t-test Kinerja
t_stat_Kinerja, p_value_Kinerja = ttest_rel(data['Kinerja Ekspektasi'], data['Kinerja Realita'])
```

```python
t_stat_Manfaat, p_value_Manfaat
```

```python
t_stat_Kontribusi, p_value_Kontribusi
```

```python
t_stat_Kinerja, p_value_Kinerja
```
